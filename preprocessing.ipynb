```python
import pandas as pd
```

```python
player_totals = pd.read_csv('Data/Player Totals.csv')
award_shares = pd.read_csv('Data/Player Award Shares.csv')
```

```python
player_totals
```

```python
award_shares
```

```python
award_shares = award_shares.pivot_table(index=['seas_id', 'player_id'], columns='award', values='share', aggfunc='sum', fill_value=0)
```

```python
complete = player_totals.merge(award_shares, on=['player_id', 'seas_id'], how='outer', suffixes=('', '_drop'))
```

```python
complete
```

```python
# drop all columns that have the suffix '_drop'
complete = complete[[col for col in complete.columns if not col.endswith('_drop')]]
```

```python
complete.columns
```

```python
complete
```

```python
complete
```

```python
#replace all NaN values with 0
complete = complete.fillna(0)
```

```python
complete
```

```python
#restrict data to 2004 and later
complete = complete[complete['season'] >= 2004]
```

```python
complete
```

```python
# Dropping Rows Where the same Player's Data, in the same Season, was Included Multiple Times 
# Example: If Player 1 was on Team A and then traded to Team B during the season, dataset contained three records for the same player, 
# during the same season --> (Player 1 Stats for Team A, Player 1 Stats for Team B, AND Player 1 Stats for Whole Season ('tm' = 'TOT')
complete_clean = complete.drop_duplicates(subset=['season', 'player'], keep='first', ignore_index=True)
complete_clean
```

```python
# Dropping Award-Shares Variables NOT Being Used in our Project
unused_awards = ['aba mvp', 'aba roy', 'clutch_poy']
complete_clean = complete_clean.drop(unused_awards, axis=1)
```

```python
# Dropping Numerical Variable NOT Necessary for our Project
unused_num = ['birth_year']
complete_clean = complete_clean.drop(unused_num, axis=1)
```

```python
# Categorical Variables
non_int_mask = complete_clean.dtypes == object
complete_clean[complete_clean.columns[non_int_mask]]
```

```python
# Dropping League and Team Variables
complete_clean = complete_clean.drop(['lg', 'tm'], axis=1)
```

```python
# Unique Values in Position Varibale
complete_clean['pos'].value_counts()
```

```python
# Assigning Each Player (Record) to Primary Position
# Example: SG-PG --> SG and PG-SG --> PG
complete_clean['pos'] = complete_clean['pos'].map(lambda x: x[0:x.find('-')] if (x.find('-') > -1) else x) 
```

```python
# Unique Values in Position Varibale AFTER Assigning Each Record to Primary Position
complete_clean['pos'].value_counts()
```

```python
# One-Hot Encoding the Postion Variable
complete_clean = pd.get_dummies(complete_clean, columns=['pos'])
```
