```python
import pandas as pd
```

```python
player_totals = pd.read_csv('Data/Player Totals.csv')
award_shares = pd.read_csv('Data/Player Award Shares.csv')
```

```python
player_totals
```

```python
award_shares
```

```python
award_shares = award_shares.pivot_table(index=['seas_id', 'player_id'], columns='award', values='share', aggfunc='sum', fill_value=0)
```

```python
complete = player_totals.merge(award_shares, on=['player_id', 'seas_id'], how='outer', suffixes=('', '_drop'))
```

```python
complete
```

```python
# drop all columns that have the suffix '_drop'
complete = complete[[col for col in complete.columns if not col.endswith('_drop')]]
```

```python
complete.columns
```

```python
complete
```

```python
complete
```

```python
#replace all NaN values with 0
complete = complete.fillna(0)
```

```python
complete
```

```python
#restrict data to 2004 and later
complete = complete[complete['season'] >= 2004]
```

```python
complete
```

```python
complete.to_csv('base.csv', index=False)
```
